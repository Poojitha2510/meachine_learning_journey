```python
#logical operators: can work for any two data types 
""" 
=> and, or binary operators
 => these operators are used to combine multiple
 => and: if both the operands are true then only the result is true otherwise false
 ........> if the operand is true then result is the value is second operand
 ........> if the operand is false then result is the value is first operand
 example:
 print(10 and 20) # 20
 print(0 and 20) # 0

 => or: if any one of the operands is true then the result is true otherwise false
........> if the operand is true then result is the value is first operand 
........> if the operand is false then result is the value is second operand
 example: print(10 or 20) # 10 
          print(0 or 20) # 20

 => not: it is a unary operator that inverts the boolean value of an operand
 example:
   print(not "arvind") # False
    print(not "") # True

 Truthy:all values are considered as truthy except the following falsy values
 false,0,0.0,"",[],(),{} are considered as falsy values in python
 print(bool(0)) # False
 """
print(True and True) # True
print(True or False) # True
print(not True) # False
#between integers
print(10 and 20) # 20
print(" " and "")
print(1 and 3)
```

```python
#conditional statements
"""
if statement:=> it is used to execute a block of code if a specified condition is true
example:
age = 18
if age >= 18: 
    print("you are eligible to vote")
elif statement:=> Can be used right after another 'if' statement or another 'elif' statement
else statement:=> it is used after if or elif statement to execute a block of code when the condition in if and elif statements are false

=> In an if,elif,else statement, only one block of code will be executed based on the first true condition. If all conditions are false, then the else block will be executed (if it is present).
"""
if 10 > 5:
    print("10 is greater than 5")
    print("this is if block")
print("this is outside of if block")

#elif statement
num=32
if num%2==0:
    print("the number is even")
elif num%2!=0:
    print("the number is odd")
else:
    print("the number is zero")
#example2
gender="m"# m , f, o
if gender=="m":
    print("male")
elif gender=="f":
    print("female")
else:
    print("other")


```

```python
#loops: => it is used to execute a block of code repeatedly until a specified condition is true
"""
for, while

=>while loop: it is used to execute a block of code repeatedly until a specified condition is false
      .......> syntax:while condition: # block of code to be executed
      example:
      num=0
      while num<5:
          print(num)
          num+=1
=>for loop: we have to give limited range(SEQUENCE OF NUMBERS) of values to execute a block of code repeatedly
  .....>range(0,5,1) => 0,1,2,3,4 => 0 is starting point, 5 is ending point, 1 is step size(end value always exclusive)
        syntax: for variable in sequence: # block of code to be executed
        example:
        for i in range(0,5,1):
            print(i)
"""
#while loop example
num=0
while num<5:
    print(num)
    num+=1
print("progrm ended",num)
#range function
print(1 in range(0,5,1)) # True
print(5 in range(0,5,2)) # False
#for loop example
for i in range(0,5,3):# i is variable
    print(i)
#ex2
for i in "python":# strings are also considered as sequence of characters
    print(i)
```
