```python
"""
Lists:
Primitive data types:int,float,boolean,string
Non-primitive data types:we can organize multiple data in a single variable:lists,tuples,sets,dictionaries
names= ["Alice","Bob","Charlie"]
=>lists in python are ordered and mutable(can be modified).
=>lists in python are the heterogeneous data type(we can store different types of data in a list)
"""
names= ["Alice","Bob","Charlie"]
print(type(names))#list
#to print lenth of names
print(len(names))#3
#index:means the position of an element in a list,starting from 0
names= ["Alice","Bob","Charlie"]
#         0       1         2 =>positive index  (in python we have negative indexing as well and it stsrts with -1 to avoid the conflict)
print(names[-2],names[1])#Charlie Charlie
#to modify the list we can use index (update the value at index 1)
names[1]="David"
print(names)#['Alice', 'David', 'Charlie']
print(names[-2],names[1])

"""a=10
a=12
(updating the entire value of a variable)"""
a=[3,4,5]
a[1]=10#updating the value at index 1 not the entire list
#lists are heterogeneous data type
my_list=[1,"hello",3.14,True,[0,9,8],[0,3,2]]
print(len(my_list))#6
print(my_list[4])#[0, 9, 8]
print(my_list[4][-1])#8

#Lists are mutable :so we can add,delete,update elements from the list
num=[1,2,3,4,5]
num.append(0)# add elements to the list at the end of the list

# how to add something in between the list
num.insert(2,10)# add 10 at index 2
# to delete an element from the list we can use pop() method
num.pop()# it will remove the last element from the list
num.pop(2)# it will remove the element at index 2
#to print the deleted value we can store the popped value in a variable
deleted_value=num.pop(2)
print(deleted_value)#10
#to sort the list in ascending order we can use sort() method
num.sort()

""" Iteration throuh a list: we can use for loop to iterate through a list and access each element of the list
sequence data type: "Arvind",[3,4,5],range(1,10)"""

name="Arvind"
#for in loop
for char in name:#char ='A,'r','v','i','n','d'
    print(char)


name=["Arvind","Rajesh","Suresh"]
#for in loop
for char in name:#char ='Arvind','Rajesh','Suresh'
    print(char)


index=0
while index<len(name):
    print(name[index])
    index+=1


```

```python
""" Tuples: """
# Tuples are ordered and immutable (cannot be changed) data structures in Python
# They are similar to lists but use parentheses instead of square brackets
#common:indexing,iteration(as touple is also in sequence data type),heterogeneous data type

scores=(85,90,78,92)
#        0   1  2 3=> positive indexing
#        -4  -3 -2 -1 =>negative indexing
print(type(scores))#tuple
print(scores[1])#90
print(scores[-1])#92
#to modify the tuple we cannot use index as tuples are immutable
# we cant modify the tuple but we can create a new tuple by concatenating the existing tuple with a new tuple
```

```python
""" Sequence data types: string,list,touple.
Slicing(sub sequence): we can extract a portion of a sequence data type using slicing.
Example: "abcd" =>slicing =>"abc","bcd","a","d"


value="abcd"
value[start:end:step] start=>inclusive index, end=>exclusive index, step=>the step size for slicing
"""

name="Arvind"
#to slice the name from index 0 to 4 with step size 1
print(name[3:6:1])

nums=[1,2,3,4,5,6,7,8,9]
print(nums[1:8:2])
print(nums[::2])# it will print all the elements with step size 2
print(nums[::-1])# it will print the list in reverse order
print(nums[5:2:-1])# it will print the elements from index 5 to 3 in reverse order

str="Arvind"
print(str[-1:-4:-2])
print(str[-1:4:1])

#Default values in slicing : [start:end:step] => step>0 => start=0, end=len(sequence) (for  positive step size)
#Default values in slicing : [start:end:step] => step<0 => start=len(sequence)-1, end=-1 (for negative step size)
nums=[4,5,9,10,True]
print(nums[::])# it will print the entire list as we are using default values for start,end and step
print(nums[::-2])
```

```python
""" Dictionaries: Key,value pair data structure in python
Dictionaries are unordered and mutable data structures in Python
=>not in sequence data type as it is not ordered data structure
=>we can access the value of a dictionary using the key
=>duplicates are not allowed in dictionaries as keys are unique
=>we can use the get() method to access the value of a dictionary using the key
"""

user={"name":["Aravind","Rajesh","mom"],"age":25,"city":"Delhi"}
print(type(user))#dict
print(user["name"],user.get("age"))#Aravind 25
#to update the age value in the dictionary we can use the key
user["age"]=30
print(user["age"])
# to add a new key value pair to the dictionary we can use the key
user["country"]="India" # it is one way to add the key value pair to the dictionary
user.update({"gender":"Male"})# it is another way to add the key value pair to the dictionary
user.setdefault("hobby","coding")# it is another way to add the key value pair to the dictionary
print(user)

# to delete the key value pair from the dictionary we can use the pop() method
user.pop("city1","it is not found")# it will remove the key value pair with key "city"
deleted_value=user.pop("age")# it will remove the key value pair with key "age" and return the value of the deleted key
print(user,deleted_value)

# iteration in a dictionary w use for in loop

for key in user:
    print(key,user[key])

print(user["name"].pop(0))
print(user["name"])
```
